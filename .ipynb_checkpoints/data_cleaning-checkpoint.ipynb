```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/PlayingNumbers/ds_salary_proj/master/glassdoor_jobs.csv",index_col=0)
```

```python
df
```

```python
"""
salary parsing(removing minus one)
Company name text only
state field
age of company 
parsing of job description (python, etc.)
"""
```

```python
df = df[df['Salary Estimate'] != '-1']
df
```

```python
# Removing the text field in Glassdoor Estimate
salary = df['Salary Estimate'].apply(lambda x: x.split('(')[0])
salary
```

```python
# Removing the K and the dollar sign
minus_kd = salary.apply(lambda x: x.replace('K','').replace('$', ''))
minus_kd
```

```python
# Appending Salary Estimate to the Data Frame
df['hourly'] = df['Salary Estimate'].apply(lambda x: 1 if 'per hour' in x.lower() else 0)
df
```

```python
df['employer_provided'] = df['Salary Estimate'].apply(lambda x: 1 if 'employer provided salary:' in x.lower() else 0)
df
```

```python
min_hr = minus_kd.apply(lambda x: x.replace('k', '').replace('$', '').replace('employer provided salary:', ''))
min_hr
```

```python
df['min_salary'] = min_hr.apply(lambda x: x.split('-')[0])      
df
```

```python
df['max_salary'] = min_hr.apply(lambda x: x.split('-')[1])      
df
```

```python
df['avg_salary'] = (df['min_salary']+df['max_salary'])
df
```

```python
""""
Company name text only
State field
Age of Company
Parsing of job description (python, etc.)
""""
```

```python
df
```

```python
# Company Name text field
df['company_text'] = df.apply(lambda x: x['Company Name'] if x['Rating'] < 0 else x['Company Name'][:-1], axis = 1)
df
```

```python
# State field
df['job_state'] = df['Location'].apply(lambda x: x.split(',')[1])
df
```

```python
df['Company_Name'] = df['Company Name'].apply(lambda x: x.replace('\n',''))
df
```

```python
df['same_state'] = df.apply(lambda x: 1 if x.Location == x.Headquarters else 0, axis = 1)
df
```

```python
#age of company
df['age'] = df.Founded.apply(lambda x: x if x <1 else 2023 - x)
df
```

```python
#parsing of job description (python, etc.)
df['python_yn'] = df['Job Description'].apply(lambda x: 1 if 'python' in x.lower() else 0)
df.head()
```
